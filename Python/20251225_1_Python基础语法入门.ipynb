```python
#定义变量
molecule_name = "Benzene"
molecule_weight = 78.11
carbon_atoms = 6
is_aromatic = True          #True and False是布尔值，属于内置关键字，不需要引号
#打印变量
print(molecule_name)
print("分子量:",molecule_weight)
print("碳原子数:",carbon_atoms)
print("是否芳香族:",is_aromatic)
```

Python的基本数据类型
字符串     str      "Bezene"        分子式、SMILES
浮点数     float    78.11
整数       int      6
布尔值     bool     True/False      判断条件

```python
#数学运算
#1.计算分子量：以C2H5OH为例
C = 12.01
H = 1.008
O = 16.00
ethanol_weight = 2*C + 6*H + 1*O
print(ethanol_weight)
```

```python
#2.计算摩尔浓度：已知10gNACL溶于500MLH2O
nacl_mass = 10
nacl_molar_mass = 58.5
volume = 0.5

moles = nacl_mass / nacl_molar_mass
concentration = moles / volume

print(f"{moles:.2f}mol")
print(f"{concentration:.2f}mol/L")   #字符串格式化使用f-string    #:.2f表示保留两位小数
```

```python
#字符串格式化：使用f-string
compound = "COF-5"
yield_percent = 85
print(f"{compound}的产率为{yield_percent}%")
```
