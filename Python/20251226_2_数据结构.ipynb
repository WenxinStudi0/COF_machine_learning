```python
##列表：列表是有序的、可变的元素集合，用[]表示
#创建列表
molecules = ["benzene","toluene","aniline","phenol"]
print(molecules)

molecular_weights = [78.11,92.14,93.13,94.11]
print(molecular_weights)         #列表可以包含不同类型的数据，之前我们学过基本的数据类型：字符串、整数、浮点数、布尔值
```

```python
#访问列表元素
molecules = ["benzene","toluene","aniline","phenol"]
        #索引代表：0，       1，      2，       3

print(molecules[0])  #python索引从0第一个，是benzene
print(molecules[-1]) #python索引倒数第一个，是phenol

print(molecules[1:3]) #表示到3但不包含3，即toluene,aniline
print(molecules[:2])  #表示从头开始但不包含2，即benzene,toluene  
print(molecules[2:])  #表示从2开始直到结束，即aniline,phenol   
```

```python
#列表常用操作

#添加元素
molecules = ["benzene","toluene"]
molecules.append("aniline") #append(新元素)表示在末尾添加新元素
print(molecules)

#插入元素
molecules.insert(1,"xylene") #insert(位置，新元素)表示在位置插入新元素
print(molecules)

#删除元素
#1.删除指定元素
molecules.remove("xylene")
print(molecules)
#2.删除索引元素
del molecules[0]
print(molecules)

#列表长度
print(len(molecules))

#检查元素是否存在
if "toluene" in molecules:
    print("yes!")
```

```python
#列表练习题目：不同温度下的产率，寻找最佳反应温度
temperatures = [80,100,120,140,160]
yields = [45,68,82,75,60]

max_yield = max(yields) #寻找最大产率值，即82
best_temp_index = yields.index(max_yield)
#yields.index(max_yield)表示在产率中查找最大产率值即82的索引位置，即2
#best_temp_index = yields.index(max_yield)表示最佳温度所在的索引位置，即2
#作用：寻找位置
best_temp = temperatures[best_temp_index]
#best_temp = temperatures[best_temp_index]用这个位置去温度列表中读取最佳反应温度
#作用：利用位置寻找温度

print(f"最高产率：{max_yield}%")
print(f"最佳反应温度：{best_temp}℃")
```

```python
##字典：是无序的键值对集合，用{}表示
#创建字典：存储一个分子的完整信息
benzene = {
    "name":"benzene",
    "formula":"C6H6",
    "molecular_weight":78.11,
    "boiling_point":80.1,
    "is_aromatic":True}
print(benzene)
```

```python
#访问字典元素

#通过键访问值
print(benzene["name"]) #benzene

#使用get()方法:更安全，不存在时返回None
print(benzene.get("formula")) #formula
print(benzene.get("density")) #None

#修改值
benzene["boiling_point"] = 80.2
print(benzene["boiling_point"]) #80.2

#添加新键值对
benzene["melting_point"] = 5.5
print(benzene)
```

```python
#字典常用操作
compound = {
    "name":"COF-5",
    "BET_surface_area":1670,
    "pore_size":2.7}

#获取所有键
print(compound.keys())

#获取所有值
print(compound.values())

#获取所有键值对
print(compound.items())

#检查键是否存在
if "BET_surface_area" in compound:
    print("yes!")

#删除键值对
del compound["pore_size"]
print(compound)
```

```python
#字典练习题目：管理多个化合物数据，筛选高比表面积的COF
cof_database = [
    {"name":"COF-1","BET":750,"yield":65},
    {"name":"COF-5","BET":1670,"yield":80},
    {"name":"COF-10","BET":1260,"yield":72}]

for cof in cof_database:
    if cof["BET"] >1000:
        print(f"{cof["name"]}具有高比表面积:{cof["BET"]}")
```

```python
##元组：存储不应改变的数据，用()
molecule_info =("benzene","C6H6",78.11)
print(molecule_info[0])
```

```python
##集合：寻找元素，用{}
elements ={"C","H","O","N","C","H"} #重复会被自动去除
print(elements) #{"C","H","O","N"}

#应用
molecule1_elements = {"C","H","O"}
molecule2_elements = {"C","H","N"}
#1.并集
all_elements = molecule1_elements | molecule2_elements
print(all_elements)
#2.交集
common_elements = molecule1_elements & molecule2_elements
print(common_elements)
#3.差集：仅在molecule1中的元素
unique_to_mol1 = molecule1_elements - molecule2_elements
print(unique_to_mol1)
```

#数据结构对比与选择
数据结构       特点        符号       应用场景
列表       有序 可重复      []       实验数据
字典         键值对         {}       分子属性
元组       有序 不可变      ()       常数数据
集合        无序 唯一       {}       元素种类

```python
##综合练习
#练习1.创建COF单体数据库
monomers = [
    {"name": "1,4-Benzenediamine","formula": "C6H8N2","mw": 108.14,"type": "amine"},
    {"name": "Terephthalaldehyde","formula": "C8H6O2","mw": 134.13,"type": "aldehyde"},
    {"name": "1,3,5-Triformylbenzene","formula": "C9H6O3","mw": 162.14,"type": "aldehyde"}]
#任务1：打印所有醛类单体名称
print("醛类单体:")
for monomer in monomers:
    if monomer["type"] == "aldehyde": # "="是赋值，"=="是比较，判断左右两边是否相等
        print(f"{monomer["name"]}")
#任务2.计算所有单体平均分子量
total_mw = sum(m["mw"] for m in monomers)
avg_mw = total_mw / len(monomers)
print(f"\n平均分子量: {avg_mw:.2f} g/mol")
```

```python
#练习2：实验数据分析
experiment_data = {"time_hours": [6, 12, 18, 24, 30],"yield_percent": [35, 58, 75, 82, 78]}
# 任务：找出最佳反应时间
max_yield_index = experiment_data["yield_percent"].index(max(experiment_data["yield_percent"]))
optimal_time = experiment_data["time_hours"][max_yield_index]
optimal_yield = experiment_data["yield_percent"][max_yield_index]

print(f"最佳反应时间: {optimal_time}小时")
print(f"对应产率: {optimal_yield}%")
```
